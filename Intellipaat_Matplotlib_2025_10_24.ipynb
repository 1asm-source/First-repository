## Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars_df = pd.read_csv('mtcars2.csv', delimiter=',')
cars_df.shape
```

```python
# drop S.No col
cars_df.drop(columns=['S.No'], inplace=True)

# rename 'Unnamed: 1' to 'model'
cars_df.rename(columns={'Unnamed: 1': 'model'}, inplace=True)

# fill null values
cars_df['qsec'].fillna(cars_df['qsec'].mean(), inplace=True)
```

```python
cars_df
```

#### Line Plot

```python
plt.figure(figsize=(5,3))
plt.plot(cars_df['hp'])
plt.plot(cars_df['disp'])
plt.show()
```

Add a legend

```python
plt.figure(figsize=(5,3))
plt.plot(cars_df['hp'], label='HP')
plt.plot(cars_df['disp'], label='Disp')
plt.legend(loc='best')
plt.show()
```

**Stack Plot**

```python
plt.stackplot
```

```python
plt.figure(figsize=(5,3))

x = cars_df['disp'].index
y = cars_df['disp']

plt.stackplot(x,y)
plt.show()
```

```python
plt.figure(figsize=(5,3))

x = cars_df['disp'].index
y1 = cars_df['disp']
y2 = cars_df['hp']

plt.stackplot(x,y1, labels='Disp')
plt.stackplot(x,y2, labels='HP')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(5,3))

x = cars_df['disp'].index
y1 = cars_df['disp']
y2 = cars_df['hp']

plt.stackplot(x,y2, labels='HP')
plt.stackplot(x,y1, labels='Disp')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(5,3))

x = cars_df['disp'].index
y1 = cars_df['disp']
y2 = cars_df['hp']

plt.stackplot(x,y2, labels='HP', alpha=0.7, color='#08D6C8')
plt.stackplot(x,y1, labels='Disp', alpha=0.6, color='#DF8A08')
plt.legend(loc='best')
plt.show()
```

**Bar Chart**

```python
plt.figure(figsize=(5,3))

x = cars_df['cyl'].index
y = cars_df['cyl']

plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(5,3))

x = cars_df['model']
y = cars_df['cyl']

plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,3))

x = cars_df['model']
y = cars_df['cyl']

plt.bar(x,y)

plt.xticks(rotation=90)

plt.show()
```

**Horizontal Bar chart**

```python
plt.figure(figsize=(3,8))

x = cars_df['model']
y = cars_df['cyl']

plt.barh(x,y)
plt.show()
```

**Scatter plot**

Qsec vs displacement

```python
plt.figure(figsize=(5,3))

x = cars_df['qsec']
y = cars_df['disp']

plt.scatter(x, y)
plt.show()
```

```python
plt.figure(figsize=(5,3))

x = cars_df['qsec']
y = cars_df['disp']

plt.scatter(x, y, marker='*', color='slateblue', s=50)

plt.title('Qsec vs Dsiplacement\n', fontweight='bold', fontsize=12)
plt.xlabel('\nTime taken for quarter mile')
plt.ylabel('Displacement\n')

plt.show()
```

**Subplots**

```python
plt.figure(figsize=(5,3))

plt.scatter(cars_df.qsec, cars_df.disp, marker='*', color='#1170E5', s=50)
plt.stackplot(cars_df.disp.index, cars_df.disp, labels='Disp', alpha=0.7, color='#CC6CE7')
```

```python
plt.figure(figsize=(11,3))

# subplot 1
plt.subplot(1,2,1)
plt.scatter(cars_df.qsec, cars_df.disp, marker='*', color='#1170E5', s=50)

# subplot 2
plt.subplot(1,2,2)
plt.stackplot(cars_df.disp.index, cars_df.disp, labels='Disp', alpha=0.7, color='#CC6CE7')

plt.show()
```

```python
plt.figure(figsize=(5,6))

# subplot 1
plt.subplot(2,1,1)
plt.scatter(cars_df.qsec, cars_df.disp, marker='*', color='#1170E5', s=100)

# subplot 2
plt.subplot(2,1,2)
plt.stackplot(cars_df.disp.index, cars_df.disp, labels='Disp', alpha=0.7, color='#CC6CE7')

plt.show()
```

**Histogram**

```python
plt.figure(figsize=(5,3))
plt.hist(cars_df['disp'])
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.hist(cars_df['disp'], color='slateblue', alpha=0.6, edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.hist(cars_df['cyl'])
plt.show()
```

***Box plot***

**Violin plkot**

**Pie chart**

```python
animals = ['dog', 'cat', 'wolf', 'lion']
weights = [30, 10, 50, 90]

plt.pie(weights, labels=animals, autopct='%1.0f%%')
plt.title('Pie Chart')
plt.show()
```

## Seaborn

```python
import seaborn as sns
```
